# Kreiranje uzorka slika za stacking
Generiranje nasumičnog uzorka slika u kojima je šum znatno veći od skrivene slike

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc 
uzorak = []
for i in range(10000):
    X=np.zeros((15,15))
    X[4,2:4]=1
    X[4, 9:11]=1
    X[7,5]=1
    X[12,2:11]=1
    X=X+np.random.normal(np.random.uniform(-1,1),10,15*15).reshape(15,15)
    uzorak.append(X)
plt.figure()
plt.imshow(X,vmin=0)
```

# Usporedba mean i median stackinga

```python
f,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(np.mean(uzorak,axis=0),vmin=10/np.sqrt(10000))
ax[1].imshow(np.median(uzorak,axis=0),vmin=10/np.sqrt(10000))
ax[0].set_title("mean stacking")
ax[1].set_title("median stacking")
```

# Usrednjavanje crno-bijelog filma

```python
uzorak = []
f,ax = plt.subplots(6,6,figsize=(6*5,6*5))
ax = np.array(ax).flatten()
for i in range(36):
    X=np.zeros((15,15))
    X[4,2:4]=1
    X[4, 9:11]=1
    X[7,5]=1
    X[12,2:11]=1
    X=X+np.random.normal(np.random.uniform(-1,1),1,15*15).reshape(15,15)
    uzorak.append(X)
    ax[i].imshow(X,vmin=0,cmap="Greys")
    ax[i].set_title("t=%d/30 sekundi"%i)
f.tight_layout()
```

```python
f,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(np.mean(uzorak,axis=0),vmin=10/np.sqrt(10000),cmap="Greys")
ax[1].imshow(np.median(uzorak,axis=0),vmin=10/np.sqrt(10000),cmap="Greys")
ax[0].set_title("mean stacking")
ax[1].set_title("median stacking")
```

# Usrednjavanje filma u boji

```python
uzorak = []
f,ax = plt.subplots(6,6,figsize=(6*5,6*5))
ax = np.array(ax).flatten()
for i in range(36):
    u1=[]
    for _ in range(3):
        X=np.zeros((15,15))
        X[4,2:4]=1
        X[4, 9:11]=1
        X[7,5]=1
        X[12,2:11]=1
        X=X+np.random.normal(np.random.uniform(-1,1),1,15*15).reshape(15,15)
        u1.append(X)
    ax[i].imshow(np.array(u1).T,vmin=0,cmap="Greys")
    ax[i].set_title("t=%d/30 sekundi"%i)
    uzorak.append(u1)
f.tight_layout()
```

```python
f,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(np.mean(uzorak,axis=0).T,vmin=10/np.sqrt(10000),cmap="Greys")
ax[1].imshow(np.median(uzorak,axis=0).T,vmin=10/np.sqrt(10000),cmap="Greys")
ax[0].set_title("mean stacking")
ax[1].set_title("median stacking");
```
